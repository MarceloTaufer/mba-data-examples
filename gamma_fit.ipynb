```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gamma
import math
```

```python
mean = 200.
variance = 25000.
std_dev = math.sqrt(variance)
a = mean*mean / variance
k = variance / mean
```

```python
# samples = np.random.normal(loc=mean, scale=sdt_dev, size=1000)
samples = np.random.gamma(shape=a, scale=k, size=1000)
```

```python
print(samples)
```

```python
# Compute a histogram of the sample
hist_bins = np.linspace(0, 600, 100)
histogram, bins = np.histogram(samples, bins=hist_bins, density=False)
```

```python
# Plot histogram
plt.figure(figsize=(14, 8))
plt.hist(samples, hist_bins, label="Histogram of Samples")
plt.legend()
plt.show()
```

```python
dist_bins = np.linspace(0, 600, 100)
dist_bin_centers = 0.5*(dist_bins[1:] + dist_bins[:-1])
```

```python
# fitting normal function to the samples
norm_loc, norm_scale = norm.fit(samples)
norm_pdf = norm.pdf(x=dist_bins, loc=norm_loc, scale=norm_scale)
```

```python
fig, ax1 = plt.subplots(figsize=(14, 8))

ax1.hist(samples, bins, label="Histogram of Samples", color='tab:blue')
ax1.set_xlabel('response time (s)')
ax1.set_ylabel('samples', color='tab:blue')

ax2 = ax1.twinx() 

ax2.plot(dist_bins, norm_pdf, label="Probability Density Function", color='tab:orange')
ax2.set_ylabel('probability', color='tab:orange')

fig.tight_layout()
plt.show()
```

```python
# fitting gamma function to the samples
shape, loc, scale = gamma.fit(samples)
gamma_pdf = gamma.pdf(x=dist_bins, a=shape, loc=loc, scale=scale)
```

```python
fig, ax1 = plt.subplots(figsize=(14, 8))

ax1.hist(samples, bins, label="Histogram of Samples", color='tab:blue')
ax1.set_xlabel('response time (s)')
ax1.set_ylabel('samples', color='tab:blue')

ax2 = ax1.twinx() 

ax2.plot(dist_bins, gamma_pdf, label="Probability Density Function", color='tab:orange')
ax2.set_ylabel('probability', color='tab:orange')

fig.tight_layout()
plt.show()
```
